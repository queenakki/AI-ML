```python
import numpy as np
import matplotlib.pyplot as plt

# Define membership functions
x = np.linspace(0, 4, 1000)

# Climbing Difficulty (CD)
cd_low = np.maximum(0, 1 - abs((x - 0) / 2))
cd_medium = np.maximum(0, 1 - abs((x - 2) / 2))
cd_high = np.maximum(0, 1 - abs((x - 4) / 2))

# Walking Difficulty (WD)
wd_easy = np.maximum(0, 1 - abs((x - 0) / 2))
wd_moderate = np.maximum(0, 1 - abs((x - 2) / 2))
wd_difficult = np.maximum(0, 1 - abs((x - 4) / 2))

# Terrain Type (TT)
tt_steep = np.maximum(0, 1 - abs((x - 0) / 2))
tt_downhill = np.maximum(0, 1 - abs((x - 2) / 2))
tt_smooth = np.maximum(0, 1 - abs((x - 4) / 2))

# Overall Tour Difficulty (OTD) - Fuzzy Rules
otd_easy = np.minimum(cd_low, np.minimum(wd_easy, tt_smooth))
otd_moderate = np.minimum(cd_medium, np.minimum(wd_moderate, tt_downhill))
otd_difficult = np.minimum(cd_high, np.minimum(wd_difficult, tt_steep))

# Plot the membership functions
plt.figure(figsize=(10, 6))

plt.plot(x, cd_low, label='CD Low')
plt.plot(x, cd_medium, label='CD Medium')
plt.plot(x, cd_high, label='CD High')

plt.plot(x, wd_easy, label='WD Easy')
plt.plot(x, wd_moderate, label='WD Moderate')
plt.plot(x, wd_difficult, label='WD Difficult')

plt.plot(x, tt_steep, label='TT Steep')
plt.plot(x, tt_downhill, label='TT Downhill')
plt.plot(x, tt_smooth, label='TT Smooth')

plt.plot(x, otd_easy, '--', label='OTD Easy (Rule)')
plt.plot(x, otd_moderate, '--', label='OTD Moderate (Rule)')
plt.plot(x, otd_difficult, '--', label='OTD Difficult (Rule)')

plt.fill_between(x, 0, otd_easy, alpha=0.1, color='r')
plt.fill_between(x, 0, otd_moderate, alpha=0.1, color='g')
plt.fill_between(x, 0, otd_difficult, alpha=0.1, color='b')

plt.title('Fuzzy Logic for Trekking Camp')
plt.xlabel('Input Value')
plt.ylabel('Membership Degree')
plt.legend()
plt.show()
```
